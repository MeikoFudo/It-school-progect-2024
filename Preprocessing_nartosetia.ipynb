```python
import os
import csv
from bs4 import BeautifulSoup

def extract_data_from_file(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        html_content = file.read()
    soup = BeautifulSoup(html_content, 'html.parser')
    trs = soup.find_all('tr')

    data_by_width = []
    for tr in trs:
        tds = tr.find_all('td')
        if len(tds) >= 2:
            data_by_width.append((tds[0].text.strip(), tds[1].text.strip(), os.path.basename(file_path).split('.')[0]))

    print(f"Обработанный файл: {file_path}")
    print(f"Извлечено {len(data_by_width)} пары текстов.")
    return data_by_width



all_data = []
for file in new_links:
    all_data.extend(extract_data_from_file(file))

with open('output.csv', 'w', newline='', encoding='utf-8') as csvfile:
    csv_writer = csv.writer(csvfile, delimiter='|')

    headers = ['id', 'Os_lang', 'Ru_lang', 'Source_File']
    csv_writer.writerow(headers)

    for idx, (os_text, ru_text, file_name) in enumerate(all_data):
        csv_writer.writerow([idx + 1, os_text, ru_text, file_name])

print("Извлечение данных и запись CSV завершены.")
```
